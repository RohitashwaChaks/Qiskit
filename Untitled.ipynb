```python
from qiskit import QuantumCircuit, Aer, assemble
from math import pi, sqrt
import numpy as np
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere
from qiskit_textbook.tools import array_to_latex
```

```python
def ExecutePlot(job, backend = 'statevector'):
    blochSphere = ""
    phasePlot = ""
    
    simulator = Aer.get_backend(backend+'_simulator')
    
    result = simulator.run(job, shots = 20408).result() # Do the simulation and return the result
    
    if backend != "qasm":
        out_state = result.get_statevector()
        array_to_latex(out_state, pretext="\\text{Statevector} = ")
#         print("StateVector: ",out_state) # Display the output state vector
        blochSphere = plot_bloch_multivector(out_state)
        phasePlot = plot_state_qsphere(out_state)
        
    
    counts = result.get_counts()
    print("Counts     : ",counts)
    histogram = plot_histogram(counts)
    return histogram, blochSphere, phasePlot
```

```python
circuit = QuantumCircuit(6,1)

circuit.h(0)
circuit.h(1)
circuit.x(2)

circuit.barrier()

circuit.cx(0,3)
circuit.cx(1,3)

circuit.barrier()

circuit.cx(1,4)
circuit.cx(2,4)

circuit.barrier()

circuit.cx(3,5)
circuit.cx(4,5)
circuit.ccx(3, 4, 5)
# circuit.x(5)

circuit.barrier()

circuit.measure(5,0)

#Draw Circuit
job = assemble(circuit)
circuit.draw('mpl')
```

```python
circuit = QuantumCircuit(4,1)

# Defining Inputs
circuit.h(0)
circuit.h(1)
circuit.h(2)

circuit.barrier()

circuit.cx(0,1)
circuit.cx(0,2)

circuit.barrier()

circuit.ccx(1, 2, 3)
circuit.x(3)

circuit.barrier()

circuit.measure(3,0)

# Returning inputs to their original state
circuit.barrier()

circuit.cx(0,1)
circuit.cx(0,2)
circuit.h(0)
circuit.h(1)
circuit.h(2)



#Draw Circuit
job = assemble(circuit)
circuit.draw('mpl')
```

```python
histogram, blochSphere, phasePlot = ExecutePlot(job)#,"qasm")
# histogram
phasePlot
```
